```python
class Mysampleclass:

  year = 2020

  def __init__(self, name, age, place):
    self.name = name
    self.age = age
    self.place = place

  def add_age(self):
    self.age = self.age + 1

  def display(self):
    print("Year is :", Mysampleclass.year)
    print("Name is :", self.name)
    print("Age is :", self.age)
    print("Place is:", self.place)
    print("_______________________")

  @classmethod
  def add_year(cls):
    cls.year = cls.year+1


x = Mysampleclass("Akshay", 23, "Palakkad")
y = Mysampleclass("Anugraha", 23, "Palakkad")

x.display()
y.display()

print('-----------------------------------------------------------------')

Mysampleclass.year = Mysampleclass.year+1
x.add_age()
x.display()
y.add_age()
y.display()

print('-----------------------------------------------------------------')

Mysampleclass.add_year()
x.add_age()
x.display()
y.add_age()
y.display()
```

```python
class Mysampleclass:

  year = 2020

  def __init__(self, name, age, place):
    self.name = name
    self.age = age
    self.place = place

  def add_age(self):
    self.age = self.age + 1

  def relocate(self, place):
    self.place = place

  def display(self):
    print("Year is :", Mysampleclass.year)
    print("Name is :", self.name)
    print("Age is :", self.age)
    print("Place is:", self.place)
    print("_______________________")

  @classmethod
  def add_year(cls):
    cls.year = cls.year+1

  @staticmethod
  def print_welcome():
    print("Hey welcome to our people's record")



x = Mysampleclass("Akshay", 23, "Palakkad")
y = Mysampleclass("Anugraha", 23, "Palakkad")

Mysampleclass.print_welcome()
print('_________________________________')

x.display()
y.display()

print('-----------------------------------------------------------------')

Mysampleclass.year = Mysampleclass.year+1
x.add_age()
x.display()
y.add_age()
y.display()

print('-----------------------------------------------------------------')

Mysampleclass.add_year()
x.add_age()
x.display()
y.add_age()
y.display()

print('-----------------------------------------------------------------')

Mysampleclass.add_year()
x.relocate("Delhi")
x.add_age()
x.display()
y.relocate("Africa")
y.add_age()
y.display()
```
